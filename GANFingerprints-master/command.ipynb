###**3X image creation(interpolation)**

```python
cd /content/drive/MyDrive/CS550/GANFingerprints-master/
```

```python
cd classifier_visNet/GANClassifier/
```

```python
from PIL import Image
import os
def enlarge(inp_directory, out_directory):
  if not os.path.isdir(out_directory):
    os.mkdir(out_directory)
  for file in os.listdir(inp_directory):
    img = Image.open(os.path.join(inp_directory, file))
    img = img.resize((img.size[0]*3, img.size[1]*3))
    name = out_directory + '/' + file[:-4] + '_3x.png'
    print(name)
    img.save(name)

root_dir = '/content/drive/MyDrive/CS550/GANFingerprints-master/classifier_visNet/GANClassifier/'
enlarge(root_dir + 'progan_original', root_dir + 'progan_3x_interpolation')
enlarge(root_dir + 'sngan_original', root_dir + 'sngan_3x_interpolation')
```
